```python
def convert_to_fixed_width(input_file, output_file, column_widths):
    
    with open(input_filepath, 'r') as input_f:
        with open(output_filepath, 'w') as output_f:
            for line in input_f:
                
                fixed_line = ''
                for column_name, width in column_widths.items():
                    
                    print(column_name, width)
                    # Extract data from the line based on column width
                    data = line[:width].ljust(width)
                    #print(data)
                    # Append the extracted data to the fixed-width line
                    fixed_line += data
                    #print(fixed_line)
                    # Remove the processed data from the line
                    line = line[width:]
                    #print(line)

                # Write the fixed-width line to the output file
                output_f.write(fixed_line + '\n')
```
